```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandlebrot(re, im, iter_num):
    z = 0.0j
    c = complex(re, im)
    
    for i in range(iter_num):
        
        z = z*z + c
        
        if(z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return iter_num
```

```python
img = np.zeros([2000, 2000])

for row, re in enumerate(np.linspace(-2, 1, 2000)):
    for col, im in enumerate(np.linspace(-1, 1, 2000)):
        img[row, col] = mandlebrot(re, im, 300)
    

result = img.T    
plt.figure(dpi = 100)
plt.imshow(result, cmap = 'hot', interpolation = 'bilinear', extent = [-2, 1, -1, 1])
plt.savefig('results/mandlebrotset.png')
```
